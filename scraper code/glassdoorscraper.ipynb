```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException, ElementClickInterceptedException
import time
import pandas as pd
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException, ElementClickInterceptedException
import pandas as pd
import time

def get_jobs(keyword, num_jobs, verbose=False):
    '''Scrapes job postings from Glassdoor and returns a pandas DataFrame.'''
    
    # Initialize the webdriver
    options = webdriver.ChromeOptions()
    options.add_argument('start-maximized')  # Start Chrome maximized
    driver = webdriver.Chrome(options=options)
    driver.set_window_size(1120, 1000)
    
    # Construct the Glassdoor search URL
    url = f"https://www.glassdoor.com/Job/jobs.htm?sc.keyword=%22data%20scientist%22&locT=C&locId=1147401&locKeyword=San%20Francisco,%20CA&jobType=all&fromAge=-1&minSalary=0&includeNoSalaryJobs=true&radius=100&cityId=-1&minRating=0.0&industryId=-1&sgocId=-1&seniorityType=all&companyId=-1&employerSizes=0&applicationType=0&remoteWorkType=0"
    driver.get(url)
    jobs = []

    while len(jobs) < num_jobs:
        time.sleep(4)  # Allow time for the page to load

        # Handle pop-up (if it appears)
        try:
            driver.find_element(By.CLASS_NAME, "ModalStyle__xBtn___29PT9").click()  # Close the modal
        except NoSuchElementException:
            pass

        # Find all job listings on the current page
        job_cards = driver.find_elements(By.CLASS_NAME, "react-job-listing")
        for job_card in job_cards:
            if len(jobs) >= num_jobs:
                break
            
            job_card.click()  # Open the job card
            time.sleep(2)

            # Collect job details
            try:
                company_name = driver.find_element(By.CSS_SELECTOR, ".css-87uc0g .e1tk4kwz1").text
                location = driver.find_element(By.CSS_SELECTOR, ".css-56kyx5 .e1tk4kwz5").text
                job_title = driver.find_element(By.CSS_SELECTOR, ".e1tk4kwz2 .css-1vg6q84").text
                job_description = driver.find_element(By.CSS_SELECTOR, ".jobDescriptionContent").text

                try:
                    salary_estimate = driver.find_element(By.CSS_SELECTOR, ".css-1xe2xww .css-1imh2hq").text
                except NoSuchElementException:
                    salary_estimate = "Not provided"

                try:
                    rating = driver.find_element(By.CSS_SELECTOR, ".css-1m5m32b .css-1imk8zj").text
                except NoSuchElementException:
                    rating = "Not provided"

                # Collect company-specific details
                headquarters = size = founded = ownership = industry = sector = revenue = "Not provided"
                try:
                    driver.find_element(By.XPATH, "//div[@data-tab-type='overview']").click()
                    time.sleep(1)
                    headquarters = driver.find_element(By.XPATH, "//label[text()='Headquarters']/following-sibling::span").text
                    size = driver.find_element(By.XPATH, "//label[text()='Size']/following-sibling::span").text
                    founded = driver.find_element(By.XPATH, "//label[text()='Founded']/following-sibling::span").text
                    ownership = driver.find_element(By.XPATH, "//label[text()='Type']/following-sibling::span").text
                    industry = driver.find_element(By.XPATH, "//label[text()='Industry']/following-sibling::span").text
                    sector = driver.find_element(By.XPATH, "//label[text()='Sector']/following-sibling::span").text
                    revenue = driver.find_element(By.XPATH, "//label[text()='Revenue']/following-sibling::span").text
                except NoSuchElementException:
                    pass

                # Append job data
                jobs.append({
                    "Job Title": job_title,
                    "Salary Estimate": salary_estimate,
                    "Job Description": job_description,
                    "Rating": rating,
                    "Company Name": company_name,
                    "Location": location,
                    "Headquarters": headquarters,
                    "Size": size,
                    "Founded": founded,
                    "Type of Ownership": ownership,
                    "Industry": industry,
                    "Sector": sector,
                    "Revenue": revenue
                })

                if verbose:
                    print(f"Job Title: {job_title}")
                    print(f"Salary Estimate: {salary_estimate}")
                    print(f"Job Description: {job_description[:200]}...")  # Truncate for readability
                    print(f"Rating: {rating}")
                    print(f"Company Name: {company_name}")
                    print(f"Location: {location}")
                    print(f"Headquarters: {headquarters}")
                    print(f"Size: {size}")
                    print(f"Founded: {founded}")
                    print(f"Ownership: {ownership}")
                    print(f"Industry: {industry}")
                    print(f"Sector: {sector}")
                    print(f"Revenue: {revenue}")
                    print("@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@")

            except Exception as e:
                print(f"Failed to collect job data: {e}")
                continue

        # Navigate to the next page
        try:
            driver.find_element(By.XPATH, "//li[@class='next']//a").click()
        except NoSuchElementException:
            print(f"Scraping terminated. Reached {len(jobs)} jobs.")
            break

    driver.quit()
    return pd.DataFrame(jobs)
```

```python
jobs_df = get_jobs(
    keyword="data scientist",  # Job title or keyword to search for
    num_jobs=50,              # Number of jobs to scrape
    verbose=True              # Print progress while scraping
)

# Save the results to a CSV file
jobs_df.to_csv("glassdoor_jobs.csv", index=False)
print("Jobs saved to 'glassdoor_jobs.csv'")
```
